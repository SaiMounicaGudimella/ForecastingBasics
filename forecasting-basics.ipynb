```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Example of an increasing trend

```python
years = [1965, 1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020]
prices = [1.00, 1.20, 1.40, 1.60, 1.80, 2.00, 2.20, 2.40, 2.60, 2.80, 3.00, 3.20]
```

```python
data = pd.DataFrame({'year':years,'prices':prices})
data.head()
```

```python
ax = data.plot.line(x='year')
ax.set_title('Coffee prices over time', fontsize=16)
plt.show()
```

### Example of seasonality

```python
seasons = ["Spring 2018", "Summer 2018", "Autumn 2018", "Winter 2018",
           "Spring 2019", "Summer 2019", "Autumn 2019", "Winter 2019",
           "Spring 2020", "Summer 2020", "Autumn 2020", "Winter 2020"]
prices = [2.80, 2.60, 3.00, 3.20,
          2.80, 2.60, 3.00, 3.20,
          2.80, 2.60, 3.00, 3.20]
```

```python
data = pd.DataFrame({'Seasons':seasons,'Prices':prices})
data.head()
```

```python
ax = data.plot.line(x='Seasons')
ax.set_title('Hot Chocolate Price Over Time', fontsize=16)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
plt.show()
```

### Supervised ML models

```python
quarters = ["Q1 2018", "Q2 2018", "Q3 2018", "Q4 2018",
            "Q1 2019", "Q2 2019", "Q3 2019", "Q4 2019",
            "Q1 2020", "Q2 2020", "Q3 2020", "Q4 2020"]

sales = [48, 20, 42, 32,
         16, 58, 40, 30,
         32, 31, 53, 40]

discounts = [4,2,3,
             3,2,6,
             4,3,3,
             3,6,6]

advertising = [500,150,400,
               300,100,500,
               380,280,290,
               315,625,585]
```

```python
data = pd.DataFrame({    
    'quarter': quarters,
    'sales': sales,
    'discount': discounts,
    'advertising': advertising})
data.head()
```

```python
ax = data.plot.line(x='quarter')
ax.set_title('Sales Per Quarter', fontsize=16)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
plt.show()
```

```python
discounts_adjusted = [x*10 for x in discounts]
discounts_adjusted
advertising_adjusted = [x/10 for x in advertising]
advertising_adjusted
```

```python
data = pd.DataFrame({
    'quarter': quarters,
    'sales': sales,
    'discount': discounts_adjusted,
    'advertising': advertising_adjusted
})
```

```python
ax = data.plot.line(x='quarter')
ax.set_title('Sales Per Quarter', fontsize=16)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
plt.show()
```

```python
data.set_index('quarter', inplace=True)
```
